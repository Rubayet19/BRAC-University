```python
def present(head, element):
    while head is not None:
        if head==element:
            return True
        head=head.next
    return False

```

```python
def reverseList(head):
    newHead=None
    n=head
    while n is not head:
        newNode=n.next
        n.next=newHead
        newHead=n
        n=newNode
    return newHead



```
