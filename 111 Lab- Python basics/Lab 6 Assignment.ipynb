```python
class Calculator:  
    def __init__(self):
        print("Lets Calculate")
        
    def add(self,value,second_value):
        print("Value1:", value)
        print("Operator: +")
        print ("Value2:", second_value)
        print("Result:")
        return value+second_value
        
    def subtract(self,value,second_value):
        print("Value1:", value)
        print("Operator: -")
        print ("Value2:", second_value)
        print("Result:")
        return value-second_value
        
            
    def multiply(self,value,second_value):
        print("Value1:", value)
        print("Operator: *")
        print ("Value2:", second_value)
        print("Result:")
        return value*second_value
            
    def divide(self,value,second_value):
        print("Value1:", value)
        print("Operator: /")
        print ("Value2:", second_value)
        return "Result: "+str(value/second_value)


cal= Calculator()
value=int(input())
operator=input()
second_value=int(input())

if operator=="+":
    print(cal.add(value, second_value))
elif operator=="*":
    print(cal.multiply(value, second_value))
elif operator=="/":
    print(cal.divide(value, second_value))
else:
    print(cal.subtract(value, second_value))




```

```python
class Customer:
    def __init__(self,name):
        self.name=name
    def greet(self, name=None):
        if name==None:
            print("Hello!")
        else:
            self.name=name
            print("Hello", name)
    def purchase(self, *items):
        count=0
        for x in items:
            count+=1
        print(self.name+", you purchased",count,"item(s):" )
        for x in items:
            print(x)
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
        
```

```python
class Panda:
    def __init__(self, name, gender, age):
        self.name=name
        self.gender=gender
        self.age=age

    
    def sleep(self, num=None):
        if num==None:
            print(self.name+"'s", "duration is unknown thus should have only bamboo leaves")
            return ""
        else:
            self.num=int(num)
            if num>=3 and num<5:
                print(self.name,"sleeps",  self.num, "hours daily and should have Mixed Vegetables")
            elif num>=6 and num<=8:
                print(self.name,"sleeps",  self.num, "hours daily and should have Eggplant & Tofu")
            elif num>=9 and num<=11:
                print(self.name,"sleeps",  self.num, "hours daily and should have Broccoli Chicken")
            return ""

panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))

print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
class Cat:
    def __init__(self, color="White", action="sitting"):
        self.color=color
        self.action=action
    def changeColor(self, newColor):
        self.color=newColor
    def printCat(self):
        print("{} cat is {}".format(self.color, self.action))
        
c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
class Vehicle:
    def __init__(self):
        self.x=0
        self.y=0
    
    def moveUp(self,y=None):
        if y==None:
            self.y+=1
        else:
            self.y+=y
    def moveDown(self,y=None):
        if y==None:
            self.y-=1
        else:
            self.y-=y
    def moveLeft(self,x=None):
        if x==None:
            self.x-=1
        else:
            self.x-=x
    
    def moveRight(self,x=None):
        if x==None:
            self.x+=1
        else:
            self.x+=x
        
    def print_position(self):
        print((self.x, self.y))
    
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```
