```python
#Easy
#1
def fraction(x, y):
    if x==0 or y ==0:
        return 0
    return x/y%1
x=int(input("Enter first Number "))
y=int(input("Enter second Number "))
print(fraction(x,y))
```

```python
#2
def BMI(h,w):
    bmi=w/(h/100)**2
    return bmi
h=int(input("Enter Height "))
w=int(input("Enter Weight "))
bmi=BMI(h,w)
if bmi<18.5:
    print("Score is "+str(bmi) +". You are under weight")
    
elif bmi>18.5 and bmi<24.9:
    print("Score is "+str(bmi)+". You are Normal")
    
elif bmi>24.9 and bmi<30:
    print("Score is "+str(bmi)+". You are Overweight")
    
else:
    print("Score is "+str(bmi)+". You are Obese")
```

```python
#3
def number(minimum,maximum,divisor):
    sum=0
    for i in range(minimum, maximum,1):
        if i%divisor==0:
            sum+=i
    return sum
maximum=int(input("Enter max number "))
minimum=int(input("Enter min number "))
divisor=int(input("Enter divisor "))
print(number(minimum,maximum,divisor))
        
```

```python
#Medium
#1
def order(name, place="Mohakhali"):
    if name=="BBQ Chicken Cheese Burger":
        p=250
    if name=="Beef Burger":
        p=170
    else:
        p=200
    t=0.08*p
    if place!="Mohakhali":
        d=60
    else:
        d=40
    tp=p+d+t
    return tp
name=input("Enter your order ")
place=input("Enter location ")
print(order(name, place))
```

```python
#2
def replace_domain(email,nd,od="kaaj.com"):
    if od in email:
        count=0
        b=0   
        while email[b]!="@":
            count+=1
            b+=1
        if email[count:len(email)]=="sheba.xyz":
            print("Unchanged:",email)
        else:
            print("Changed:",email[0:count+1]+nd)
    

replace_domain(input(),input(),input())
        
```

```python
#3
def Check_Vow(name, vowels="AaEeIiOoUu"):
    count=0
    for each in name:
        if each in vowels:
            print(each,end=" ")
            count+=1
    print(".Total number of vowels:", count)
    
Check_Vow(input())
```

```python
#4
def Palindrome(word, str=""):
    for i in word:
        str=i+str
    if word==str:
        print("Palindrome")
    else:
         print("Not a Palindrome")

Palindrome(input())
```

```python
#Hard
#1
from datetime import date
def amount(name, salary, date):
    today = date.today()
    diff=today-date
    d=diff.days
    if d<1825:
        total=0.1*salary
    elif d>1825 and d<3650:
        total=0.1*salary+5000
    else:
        total=0.15*salary+15000
    print(name+":",total)
year = int(input('Enter a year '))
month = int(input('Enter a month '))
day = int(input('Enter a day '))

amount(input("Enter Name "), int(input("Enter Salary ")), date(year, month, day))

```

```python
#2
def time(days):
    years=days//365
    months=(days-years*365)//30
    day=days-years*365-months*30
    print(years,"years,",months,"months and",day,"days")
time(input())
```

```python
def capital(para):
    ls=[]
    for i in range(len(para)):
        ls.append(para[i])
    for i in range(len(para)-2):
        ls[0]=ls[0].capitalize()
        if ls[i]=='.' or ls[i]=="?" or ls[i]=="!":
            ls[i+2]=ls[i+2].capitalize()
        elif ls[i]=="i" and ls[i-1]==" " and ls[i+1]==" " or ls[i+1]=="'":
            ls[i]=ls[i].capitalize()
    for i in range(len(para)):
        print(ls[i])
capital(input())
```
